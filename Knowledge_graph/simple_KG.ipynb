```python
import igraph as ig
import matplotlib.pyplot as plt
```

```python
g=ig.Graph(directed=True)
g.add_vertices(5)  #Adding 5 vertices to the graph, g.vs object is used for vertex sequence
g.vs['fname'] = ['Adam', 'Julia', 'Martin', 'Tom', 'Rose'] #Giving each node an attribute value
g.add_edges([(0, 1), (1, 3), (4, 2), (3, 4), (2, 0)])
print(g)
```

```python
g.vs[4]['fname']
[i.index for i in g.vs.select(fname='Julia')]
```

```python
fig, ax = plt.subplots()
ig.plot(
    g, 
    target=ax,
    vertex_size=0.3,
    vertex_label=g.vs['fname'],
    vertex_color='red',
    edge_color='gray'
)
plt.show()
```

```python
# Create a graph representing a social network
g = ig.Graph(directed=False)
g.add_vertices(['Alice', 'Bob', 'Charlie', 'David', 'Eve','Alex'])
g.add_edges([('Alice', 'Bob'), ('Bob', 'Charlie'), ('Charlie', 'David'), ('David', 'Eve'), ('Eve', 'Alice'), ('Alice', 'Charlie'),('Alice','Alex')])

g.vs['label'] = g.vs['name']

# Calculate centrality
centrality = g.degree()
g.vs['size'] = [c * 10 for c in centrality]

# Plot the graph
fig, ax = plt.subplots()
ig.plot(
    g, 
    target=ax,
    vertex_size=g.vs['size'],
    vertex_label=g.vs['label'],
    vertex_color='orange',
    edge_color='black'
)
plt.show()
```
